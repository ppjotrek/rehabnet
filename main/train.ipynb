```python
import torch
import matplotlib.pyplot as plt
import numpy as np
from torchsummary import summary
from utils import load_data, emg_dataset
from model import SERes1d
from sklearn.model_selection import train_test_split
from sklearn import metrics
import time
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
training_dataset = emg_dataset("C:\\Users\\piotr\\Python\\rehabnet\\main\\data.csv")

training_dataloader = torch.utils.data.DataLoader(training_dataset, batch_size=16, shuffle=True)

eval_dataset = emg_dataset("C:\\Users\\piotr\\Python\\rehabnet\\main\\eval.csv")
eval_dataloader = torch.utils.data.DataLoader(eval_dataset, batch_size=16, shuffle=True)
'''
IDLE        3774
FD           895
FL           890
EX           884
WP           884
FH           883
FS           883
WS           883
WU           883
WR           882
HandOpen       4 -> to jest FD
'''

class_to_index_map = {"IDLE": 0, "FD": 1, "HandOpen": 1, "FH": 2, "FL": 3, "FS": 4, "EX": 5, "WR": 6, "WS": 7, "WP": 8, "WU": 9}
```

```python
model = SERes1d(8, 10, 16).to(device)
criteria = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001)
```

```python
def train(model, dataloader, criteria, optimizer, epochs=10):
    summary(model)
    model.train()
    acc_history = []
    val_history = []
    best_val_accuracy = 0
    for epoch in range(epochs):
        start = time.time()
        running_corrects = 0
        predictions = 0
        print(f"Epoch {epoch+1}/{epochs}")
        for i, (data, target) in enumerate(dataloader):
            data, target = data.to(device), target.to(device)
            output = model(data)
            for i in range(len(output)):
                if torch.argmax(output[i]) == target[i]:
                    running_corrects += 1
            loss = criteria(output, target)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            predictions += len(data)
        accuracy = (running_corrects/predictions)*100
        acc_history.append(accuracy)
        print(f"Loss: {loss.item()}, accuracy: {accuracy}, corrects: {running_corrects}")
        #validation
        model.eval()
        running_val_corrects = 0
        predictions = 0
        for i, (data, target) in enumerate(eval_dataloader):
            data, target = data.to(device), target.to(device)
            output = model(data)
            for i in range(len(output)):
                if torch.argmax(output[i]) == target[i]:
                    running_val_corrects += 1
            predictions += len(data)
        val_accuracy = (running_val_corrects/predictions)*100
        val_history.append(val_accuracy)
        print(f"Validation accuracy: {val_accuracy}, corrects: {running_val_corrects}")
        if val_accuracy > best_val_accuracy:
            best_val_accuracy = val_accuracy
            torch.save(model.state_dict(), "C:\\Users\\piotr\\Python\\rehabnet\\state_dicts\\model_state_dict.pth")
            torch.save(optimizer.state_dict(), "C:\\Users\\piotr\\Python\\rehabnet\\state_dicts\\optimizer_state_dict.pth")
        stop = time.time()
        print(f"Epoch took: {stop-start} seconds")
        print("--------------------------------------------------")
    return acc_history, val_history
```

```python
acc_history = train(model, training_dataloader, criteria, optimizer, epochs=150)
```

```python
plt.plot(acc_history[0])
plt.plot(acc_history[1])
plt.legend(["Training", "Validation"])
```

```python
actual = []
predicted = []
for data,label in eval_dataloader:
    data = data.to(device)
    out = model(data)
    actual.extend(label.numpy())
    predicted.extend(torch.sigmoid(out).argmax(dim=1).cpu().numpy())


confusion_matrix = metrics.confusion_matrix(actual, predicted,normalize='true')
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [0, 1, 2, 3, 4, 5])


cm_display.plot()
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```
